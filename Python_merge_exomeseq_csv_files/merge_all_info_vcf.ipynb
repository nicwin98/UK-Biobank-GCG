```python
import pandas as pd
import os
import glob
import numpy as np
from os import listdir
```

### Merge all info .csv files using .concat

```python
csv_files = glob.glob('path_to_csv_files//*.{}'.format('csv'))

df_info = pd.DataFrame()
for file in csv_files:
    data = pd.read_csv(file, dtype=str)
    print(file) #nice to follow
    data.index = data[['CHROM', 'POS', 'REF', 'ALT']].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
    data=data.drop(['CHROM', 'POS', 'REF', 'ALT'],axis=1)
    data=data.drop(['INFO'], axis=1) 
    df_info = pd.concat([df_info, data], axis = 1)
```

```python
df_info.head()
```


### Replace cells starting with ./.  with NaN

```python
df_info_filtered = df_info.copy()
str_cols = df_info_filtered.columns 

for column in str_cols:
     df_info_filtered.loc[df_info_filtered[column].str.startswith('.', na=False), column] = np.nan
```

### Filter rows with only Nan 

```python
# filter rows
df_info_filtered2=df_info_filtered.dropna(subset=str_cols, how="all")
```

```python
print(df_info_filtered.shape,
      df_info_filtered2.shape) #filtered row with all NaN
```


#### Prepare table for open.cravat 

```python
# make index into a column
df = df_info_filtered2.copy()
df.reset_index(inplace=True)

#split the column "index" into four seperate coulumns
df[['CHROM', 'POS', 'REF', 'ALT']] = df["index"].apply(lambda x: pd.Series(str(x).split("_")))
df=df.drop(["index"], axis=1)
df['INFO'] = 'GT:GQ:DP:AD:VAF:PL' #adds the INFO column that we deleted at import to avoid replicates from each csv file
df.head()
```

```python
#reorder to get Chrom etc. first in the DF
cols = df.columns.tolist()
cols = cols[-5:] + cols[:-5]
df = df[cols]
df.head()
```

```python
# ===============================================
# remove , <*> from ALT
# ===============================================

df["ALT"] = df["ALT"].str.replace("[,<*>]", "", regex=True)

df.head()
```
